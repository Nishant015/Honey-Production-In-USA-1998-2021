## Honey Production In The USA (1998-2021)

### Background:-
In 2006, global concern was raised over the rapid decline in the honeybee
population, an integral component of American honey agriculture. Large
numbers of hives were lost to Colony Collapse Disorder, a phenomenon of
disappearing worker bees causing the remaining hive colony to collapse.
Speculation as to the cause of this disorder points to hive diseases and
pesticides harming the pollinators, though no overall consensus has been
reached. The U.S. used to locally produce over half the honey it consumes
per year. Now, honey mostly comes from overseas, with 350 of the 400
million pounds of honey consumed every year originating from imports. This
dataset provides insight into honey production supply and demand in
America from 1998 to 2021.

### Objective:-
To visualize how honey production has changed over the years (1998-2021) in
the United States.

# 1. Looking at how the honey production yield changed from 1998 to 2021.


```python
df1 = df.groupby(['year'])['totalprod'].sum().reset_index()
```

```python
df1
```

# Visualising df1 on line chart using seaborn

```python
plt.figure(figsize=(9,4))
sns.lineplot(data = df1, x='year', y='totalprod', marker='o', color='red')
plt.title('Honey production from 1998-2021')
plt.xlabel('Year')
plt.ylabel('Total production')
plt.grid()
```

###  Observation:- 
* The overall honey production in the US has been decresing over the years.
* Total honey production = number of colonies * yield per colony.
   

 # 2. Over time, what are the major production trends across the states?

```python
state_prod = df.groupby(["State", "year"])["totalprod"].sum().reset_index()
```

```python
state_prod
```

# Visualising state_prod on line chart using seaborn


```python
sns.lineplot(data=state_prod, x='year', y='totalprod', hue='State')
plt.legend(bbox_to_anchor=(1,1))
plt.show()
```

### Observation:-
There are some states that have much higher productions than the others but this plot is a little hard to read. Let us try plotting each state seperatly for a better understanding.

```python
df2 = sns.FacetGrid(state_prod, col='State',  col_wrap=5, height=2)
df2.map(plt.plot, 'year', 'totalprod', marker='o', color='orange')
plt.show()
```

### Observation:-
* The most honey producing states of US are - California, Florida, North Dakota and South Dakota and Montana
* The honey production in California has seen a steep decline over the years.
* Florida's total production also has been on a decline.
* South Dakota has more of less maintained its levels of production.
* North Dakota has actually an impressive increase in the honey production.

# 3. Checking if the data shows any trends in terms of the number of honeyproducing colonies and yield per colony before 2006, which was when concern over Colony Collapse Disorder spread nationwide?


## Variation of number of colonies over the years


```python
df2 = df.groupby(['year'])['numcol'].sum().reset_index()
```

```python
df2
```

# Visualizing number of colonies over the years on line chart using seaborn

```python
sns.lineplot(data=df2, x='year', y='numcol',marker='o')
plt.title('Number of colonies over the years')
plt.xlabel('Year')
plt.ylabel('Number of Colonies')
plt.grid()
plt.show()
```

## Variation of Yield per colony over the years

```python
df3 = df.groupby(['year'])['yieldpercol'].sum().reset_index()
```

```python
df3
```

# Visualizing Variation of Yield per colony over the years on line chart using seaborn

```python
sns.lineplot(data=df3, x='year', y='yieldpercol',marker='d',color='green')
plt.title('Yield per colony over the years in pounds')
plt.xlabel('Year')
plt.ylabel('Honey Yield per Colony')
plt.grid()
```

### Observation:-
* The number of colonies was decreasing before 2006. So, the yield per colony also decreased, which is why it is considered a colony collapse disorder.
* After 2006 the number of colonies started increasing but the yield per colony still decreased.
* This indicates that, it is not the number of colonies that is causing a decline in totalhoney production but the yield per colony.

# 4. Are there any patterns that can be observed between total honey production and value of production every year?


```python
df4= df.groupby(['year'])['prodvalue'].mean().reset_index()
```

```python
df4
```

# Visualizing Value of Honey Production Over the Years on a bar chart using seaborn

```python
sns.barplot(x=df4['year'],y=df4['prodvalue'], edgecolor='black')
plt.title('Value of honey production over the years')
plt.xlabel('year')
plt.ylabel('Value of honey production')
plt.xticks(rotation=90)
plt.show()
```

 ### Observation:-
 * Although honey production has declined over the years, the value of production per pound has increased over time.
 * As the supply declined, the value has added to the value of honey.
